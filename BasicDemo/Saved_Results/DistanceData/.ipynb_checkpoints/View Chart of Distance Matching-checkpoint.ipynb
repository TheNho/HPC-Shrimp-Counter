```python
save_list = []
for i in range(1, 15):
    filename = f"SaveDistance{i}.txt"
    file1 = open(filename, 'r')
    read_list = file1.readlines()
    save_list += read_list
```

```python
number_list = []
for number in save_list:
    number_list.append(float(number))
```

```python
len(number_list), number_list[0]
```

```python
import numpy as np
number_array = np.array(number_list)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,5))
plt.xlabel("Point")
plt.ylabel("Distance(Pixel)")
y = np.arange(len(number_array))
plt.grid()
plt.scatter(y, number_array, alpha=0.5, marker='o', s=2)
plt.show()
```
